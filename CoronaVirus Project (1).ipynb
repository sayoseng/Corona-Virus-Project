```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('COVID_data_prepped.csv', sep=',')
```

```python
# In this project we are performing predictive analysis on the Cumulative_deaths column using a few machine learning algs
df.head(10)
```

```python
# Performing correlation on the dataset to perceive the relationship between the variables in the data set.
# We can see a significant positive correlation between Cumulative_deaths and Cumulative_cases
df.corr()
```

```python
# This is where we begin our machine learning
y = df.Cumulative_deaths
X = df.drop(['Cumulative_deaths'],axis=1)
X.count()
```

```python
# Splitting the train/test data sets from the original data set
X_train,X_test,y_train,y_test = train_test_split(X,y)
training, testing = train_test_split(df,test_size = .2)
```

```python
# Setting up regressor
regressor = LinearRegression()
trainAlg = regressor.fit(X_train,y_train)
```

```python
regressor = RandomForestRegressor()
RFtrainAlg = regressor.fit(X_train,y_train)
```

```python
regressor = DecisionTreeRegressor()
DTtrainAlg = regressor.fit(X_train,y_train)
```

```python
#Decision Tree Algorithm
y_pred = DTtrainAlg.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
#Random Forests Algorithm
y_pred = RFtrainAlg.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
#Linear Regression
y_pred = trainAlg.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```
